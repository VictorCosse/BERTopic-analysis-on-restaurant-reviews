```python
# import bertopic
from bertopic import BERTopic

# re package utile après pour le nettoyage
import re

# import pandas
import pandas as pd

# import emoji
import emoji

# sentence transformers pour transformer les phrases en vecteurs
# util pour calculer des similarités entre vecteurs
from sentence_transformers import SentenceTransformer, util
```

```python
# import fichier
df = pd.read_csv('/content/sample_20k.csv')
```

```python
df
```

```python
df[['place_id', 'text']]
```

```python
# Supprimer les URLs (tout texte qui commence par "http")
df['text'] = df['text'].apply(lambda row: re.sub(r"http\S+", "", str(row)))

# supprime toutes les lignes du dataframe 'df' où la colonne "text" est vide
df = df.loc[(df.text != ""), :]
```

```python
# définition d'une fonction pour supprimer les emojis d'une chaîne de caractères
def remove_emoji(text):
    new_text = ''
    for character in text:
        if character not in emoji.core.emojize('').replace('_', ''):
            new_text += character
    return new_text

# application de la fonction "remove_emoji" à la colonne "text" du DataFrame "df"
df['text'] = df['text'].apply(remove_emoji)
```

```python
# vérifions les NaN
df.isnull().sum()
```

```python
# vérifions les doublons
df.duplicated().sum()
```

```python
# enregistrons les ids et commentaires au format listes
ids = df['place_id'].tolist()
docs = df['text'].tolist()
```

```python
# définition d'une fonction pour supprimer les caractères non-ASCII d'une chaîne de caractères
def remove_non_ascii(text):
    return re.sub(r'[^\x00-\x7F]+', '', text)

# application de la fonction "remove_non_ascii" à chaque élément de la liste "docs"
docs = [remove_non_ascii(text) for text in docs]
```

```python
# vérifions que nous n'avons pas perdu de commentaires en route
len(docs)
```

```python
import numpy as np

# calcul des longueurs des chaînes de caractères
lengths = [len(doc) for doc in docs]

# calcul des statistiques descriptives
print("Moyenne : ", np.mean(lengths))
print("Minimum : ", np.min(lengths))
print("Maximum : ", np.max(lengths))
print("Q1 : ", np.percentile(lengths, 25))
print("Q2 : ", np.percentile(lengths, 50))
print("Q3 : ", np.percentile(lengths, 75))
print("P80 : ", np.percentile(lengths, 80))
print("P85 : ", np.percentile(lengths, 85))
print("P90 : ", np.percentile(lengths, 90))
print("P95 : ", np.percentile(lengths, 95))
```

```python
%%time
from sklearn.feature_extraction.text import CountVectorizer
from stop_words import get_stop_words

en_stop_words = get_stop_words('en')
fr_stop_words = get_stop_words('fr')

stop_words = en_stop_words + fr_stop_words

vectorizer_model = CountVectorizer(stop_words=stop_words)
```

```python
%%time
model_embedding = SentenceTransformer('all-MiniLM-L6-v2')
corpus_embeddings = model_embedding.encode(docs)
```

```python
%%time
model = BERTopic().fit(docs, corpus_embeddings)
```

```python
topics, probabilities = model.transform(docs, corpus_embeddings)
```

```python
df_topic_freq = model.get_topic_freq()
topics_count = len(df_topic_freq) - 1
df_topic_freq
```

```python
model.visualize_barchart(top_n_topics=topics_count)
```
